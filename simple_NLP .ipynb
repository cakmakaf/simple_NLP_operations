```python
class Content():

    def __init__(self,file_name):

        with open(file_name,"r",encoding = "utf-8") as file:

            file_content = file.read()
            text_words = file_content.split()

            self.simple_text_words = list()

            for vocable in text_words:
                vocable = vocable.strip()
                vocable = vocable.strip(".")
                vocable = vocable.strip("\n")

                self.simple_text_words.append(vocable)
    def word_finder(self,detecting):

        locations = list()
        say = 1
        for vocable in self.simple_text_words:
            if (vocable == detecting ):
                locations.append(say)

            say += 1

        if (len(locations) == 0):
            print("This word is NOT existing in this file...")
        else:

            print("The word of {} has been mentioned at: \n{}".format(detecting,locations))



    def words_distribution(self):

        vocable_frequency = dict()
        vocable_set = set()

        for vocable in self.simple_text_words:
            vocable_set.add(vocable)
            if (vocable in vocable_frequency):
                vocable_frequency[vocable] += 1
            else:
                vocable_frequency[vocable] = 1
        print("Frequency of all words:")
        for i, j in vocable_frequency.items():
            print("The word of {} has been mentioned {} times in the text.".format(i, j))

        print("All Words")
        for i in vocable_set:
            print(i)

            print("---------------------------------------")


file =  Content("random_script.txt")

print("""--------------------------------------

File Operations

f. Learn the frequency of each word:

s. Detect a particular sample word: 


q. To quit press 'q' please.

""")
while True:
    operation = input("Operation:")

    if (operation == "q"):
        print("Quiting....")
        break
    elif (operation == "f"):
        file.words_distribution()
    elif (operation == "s"):
        vocable = input("Which word are you looking for?")
        file.word_finder(vocable)    
    else:
        print("Invalid Operation...")
```
